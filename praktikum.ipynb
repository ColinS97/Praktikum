````python
#3.1
# amount that should be added in order to increase brightness
brightness = 100
height= image.shape[0]
width= image.shape[1]
channels = image.shape[2]
# offset ist offsetting the rectanlge from origin
bwoffset = 50
avgoffset = 150
diffoffset = 270
image = cv.imread('CVBVP-Daten/OpenCV-01/Img01a.jpg')
for y in range(height):
    for x in range (width):
        #3.1
        if x> bwoffset and x < 50+bwoffset and y > bwoffset and y < 50+bwoffset:
            for c in range(channels):
                old = image[y,x][c]
                image [y,x][c] = old + brightness
        #3.2
        if x> avgoffset and x < 100+avgoffset and y > avgoffset and y < 100+avgoffset:
            for c in range(channels):
                #int cast are needed because if not after the addition the result is to large to be stored in the uint8 datatype
                avg = (int(image[y,x][c]) + int(image[y,x-1][c]))/2
                image [y,x][c] = avg
        #3.3
        if x> diffoffset and x < 100+diffoffset and y > diffoffset and y < 100+diffoffset:
            for c in range(channels):
                diff = int(image[y,x - 1][c]) - int(image[y,x][c])
                image [y,x][c] = diff

plt.imshow(image)
plt.show
````

````python
import cv2 as cv

def random_color(random):
    """
    Return a random color
    """
    icolor = random.randint(0, 0xFFFFFF)
    return cv.Scalar(icolor & 0xff, (icolor >> 8) & 0xff, (icolor >> 16) & 0xff)
````
